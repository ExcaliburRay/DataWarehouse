```python
import pandas as pd
wildlife_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/wildlife_record.csv')
```

```python
data_range = pd.date_range(start='2009-01-01', end='2019-12-31').tolist()
```

```python
len(data_range)
```

```python
data_range[4016]
```

```python
data_dimension = pd.DataFrame(columns = ['DateKey','Date','DateYear','DateDay','DateMonth'])
```

```python
data_dimension
```

```python
data_dimension['Date'] = data_range
```

```python
length = len(data_dimension)
data_key = []
data_year = []
data_month = []
data_day = []
date = data_dimension['Date'].astype(str)
for i in range(length):
    date_split = date[i].split('-')[0]+date[i].split('-')[1]+date[i].split('-')[2]
    #format in y/m/d
    data_key.append(date_split)
    data_year.append(date[i].split('-')[0])
    data_day.append(date[i].split('-')[1])
    data_month.append(date[i].split('-')[2])
```

```python
data_dimension['DateKey'] = data_key
data_dimension['DateYear'] = data_year
data_dimension['DateDay'] = data_day
data_dimension['DateMonth'] = data_month
```

```python
data_dimension.to_csv(r'C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/date_dimension.csv',index = None, header = True)
```

```python
specie_dimension = pd.DataFrame(columns = ['SpecieKey','AppClass','Taxon','Class','Order_Info','Family','Genus','Origin'])
specie_dimension['AppClass'] = wildlife_data['App.']
specie_dimension['Taxon'] = wildlife_data['Taxon']
specie_dimension['Class'] = wildlife_data['Class']
specie_dimension['Order_Info'] = wildlife_data['Order']
specie_dimension['Family'] = wildlife_data['Family']
specie_dimension['Genus'] = wildlife_data['Genus']
specie_dimension['Origin'] = wildlife_data['Origin']
specie_dimension.drop_duplicates(inplace = True)
specie_dimension = specie_dimension.reset_index(drop=True)
```

```python
specie_key = []
for i in range(len(specie_dimension)):
    specie_key.append(i)
specie_dimension['SpecieKey'] = specie_key
```

```python
specie_dimension.to_csv(r'C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/specie_dimension.csv',index = None, header = True)
```

```python
importer_dimension = pd.DataFrame(columns = ['ImporterKey','ImporterAbbreviation','ImporterName','ImporterActiveFlag'])
importer_dimension['ImporterAbbreviation'] = wildlife_data['Importer']
importer_dimension['ImporterName'] = wildlife_data['Short Name of Importer']
importer_dimension['ImporterActiveFlag'] = wildlife_data['Importer Active State']
importer_dimension.drop_duplicates(inplace = True)
importer_dimension = importer_dimension.reset_index(drop=True)
```

```python
importer_key = []
for i in range(len(importer_dimension)):
    importer_key.append(i)
importer_dimension['ImporterKey'] = importer_key
```

```python
importer_dimension.to_csv(r'C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/importer_dimension.csv',index = None, header = True)
```
