```python
import pandas as pd
wildlife_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/metadata/wildlife_record.csv')
area_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/metadata/area.csv')
market_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/metadata/market.csv')
specie_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/metadata/specie.csv')
term_data = pd.read_csv('C:/Users/Ray/Desktop/CMULecture19Summer/Data Warehouse/project/metadata/term.csv')
```

```python
import numpy as np
Fact_data = pd.DataFrame(columns = ['TransactionKey','DateKey','SpecieKey','MarketingKey','TermKey','ImporterKey','ExporterKey','Quantity','UnitPrice','TotalPrice'])
length = len(wildlife_data)
list_TransactionKey = []
#list_DateKey = []
list_SpecieKey = []
list_MarketingKey = []
list_TermKey = []
list_ImporterKey = []
list_ExporterKey = []
list_quantity = []
list_unitprice = []
list_totalprice = []
for i in range(length):
    list_TransactionKey.append(100000+i)
    #list_DateKey.append(wildlife_data['DateKey'][i])
    for j in range(len(specie_data)):
        if(specie_data['AppClass'][j] == wildlife_data['App.'][i] and 
           (specie_data['Taxon'][j] == wildlife_data['Taxon'][i] or (specie_data['Taxon'][j] is np.nan and wildlife_data['Taxon'][i] is np.nan))and 
           (specie_data['Class'][j] == wildlife_data['Class'][i] or (specie_data['Class'][j] is np.nan and wildlife_data['Class'][i] is np.nan))and 
           (specie_data['Order_Info'][j] == wildlife_data['Order'][i] or (specie_data['Order_Info'][j] is np.nan and wildlife_data['Order'][i] is np.nan)) and 
           (specie_data['Family'][j] == wildlife_data['Family'][i] or (specie_data['Family'][j] is np.nan and wildlife_data['Family'][i] is np.nan))and 
           (specie_data['Genus'][j] == wildlife_data['Genus'][i] or (specie_data['Genus'][j] is np.nan and wildlife_data['Genus'][i] is np.nan))and 
           (specie_data['Origin'][j] == wildlife_data['Origin'][i] or (specie_data['Origin'][j] is np.nan and wildlife_data['Origin'][i] is np.nan))):
            list_SpecieKey.append(specie_data['SpecieKey'][j])
            break
            
for i in range(length):
    for k in range(len(market_data)):
        if((market_data['Purpose'][k] == wildlife_data['Purpose'][i] or ( (pd.isnull(market_data['Purpose'][k]) and pd.isnull(wildlife_data['Purpose'][i]) ) ) ) and 
          (market_data['Source'][k] == wildlife_data['Source'][i] or ((pd.isnull(market_data['Source'][k]) and pd.isnull(wildlife_data['Source'][i]) ) ) ) and 
          (market_data['Source Description'][k] == wildlife_data['Source Description'][i] or ((pd.isnull(market_data['Source Description'][k]) and pd.isnull(wildlife_data['Source Description'][i]) ) ) ) and 
          (market_data['Purpose Description'][k] == wildlife_data['Purpose Description'][i] or ((pd.isnull(market_data['Purpose Description'][k]) and pd.isnull(wildlife_data['Purpose Description'][i]) ) ) ) and 
           market_data['ActiveLag'][k] == wildlife_data['Market Active Lag'][i] ):
            list_MarketingKey.append(market_data['MarketingKey'][k])
            break
                       
for i in range(length):
    for l in range(len(term_data)):
        if(term_data['Term'][l] == wildlife_data['Term'][i]):
            list_TermKey.append(term_data['TermKey'][l])
            break
            
for i in range(length):
    for m in range(len(area_data)):
        if( (area_data['AreaAbbreviation'][m] == wildlife_data['Importer'][i] or (pd.isnull(area_data['AreaAbbreviation'][m]) and pd.isnull(wildlife_data['Importer'][i]) ) )and 
           area_data['Area_Name'][m] == wildlife_data['Short Name of Importer'][i] and 
           area_data['ActiveLag'][m] == wildlife_data['Importer Active State'][i]):
            list_ImporterKey.append(area_data['Area_Key'][m])
            break  
            
for i in range(length):
    for n in range(len(area_data)):
        if( (area_data['AreaAbbreviation'][n] == wildlife_data['Exporter'][i] or (pd.isnull(area_data['AreaAbbreviation'][n]) and pd.isnull(wildlife_data['Exporter'][i]) ) ) and 
           area_data['Area_Name'][n] == wildlife_data['Short Name of Exporter'][i] and 
           area_data['ActiveLag'][n] == wildlife_data['Exporter Active State'][i]):
            list_ExporterKey.append(area_data['Area_Key'][n])
            break
```

```python
Fact_data['TransactionKey'] = list_TransactionKey
Fact_data['DateKey'] = wildlife_data['DateKey']
Fact_data['SpecieKey'] = list_SpecieKey
Fact_data['MarketingKey'] = list_MarketingKey
Fact_data['TermKey'] = list_TermKey
Fact_data['ImporterKey'] = list_ImporterKey
Fact_data['ExporterKey'] = list_ExporterKey
Fact_data['Quantity'] = wildlife_data['Total quantity']
Fact_data['UnitPrice'] = wildlife_data['Price']
Fact_data['TotalPrice'] = wildlife_data['Total Price']
```
