```python
wildlife_data_copy = wildlife_data
```

```python
for i in range(length):
    if(wildlife_data_copy['Year'][i]<2012):
        if(wildlife_data_copy['Source'][i] in source_out_of_date.keys()):
            wildlife_data_copy.loc[i,'Source Description'] = source_out_of_date[wildlife_data_copy['Source'][i]]
            wildlife_data_copy.loc[i,'Market Active Lag'] = 0
        if(wildlife_data_copy['Purpose'][i] in purpose_out_of_date.keys()):
            wildlife_data_copy.loc[i,'Purpose Description'] = purpose_out_of_date[wildlife_data_copy['Purpose'][i]]
            wildlife_data_copy.loc[i,'Market Active Lag'] = 0
```

```python
wildlife_record = pd.merge(wildlife_data_copy,wildlife_data)
wildlife_record.drop_duplicates(inplace = True)
wildlife_record = wildlife_record.reset_index(drop=True)
```
